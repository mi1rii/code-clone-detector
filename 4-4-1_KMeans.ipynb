# Algoritmo K-means
## Carga de datos
Se trabaja con los primeros dos atributos del conjunto de datos de Iris

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd

iris = load_iris()
data= pd.DataFrame(iris.data[:,:2], columns=['x','y'])
print(data.describe())
plt.scatter(data['x'],data['y'],color='blue')
plt.show()
```

## Implementación

```python
from sklearn.cluster import KMeans

valores=data
Algoritmo= KMeans(n_clusters=2, max_iter=20,n_init=5)

Grupos=Algoritmo.fit(valores)

valores['cluster']=Grupos.labels_
print(valores['cluster'].value_counts())
#Centroides del ciclo
print(Grupos.cluster_centers_)

plt.scatter(data['x'],data['y'],c=Grupos.labels_)
plt.scatter(Grupos.cluster_centers_[:,0],Grupos.cluster_centers_[:,1], color='red')

plt.show()
```

## ¿Cómo evaluar?
- **Silhouette**: Mide qué tan similar es un punto de datos a su propio clúster en comparación con otros clústeres. Su valor varía entre \(-1\) y \(1\).

- **Inertia**: Mide qué tan compactos y coherentes son los grupos formados por un algoritmo como K-Means. Representa la suma de las distancias al cuadrado desde cada punto de datos hasta el centroide de su clúster asignado

```python
#Coeficiente de silhouette
from sklearn import metrics
sil=metrics.silhouette_score(data, Grupos.labels_,metric='euclidean')
print("Coeficiente silhouette:",sil)

inertia=Algoritmo.inertia_
print("Inercia: ",inertia)
```

### Método del codo
Es una técnica gráfica para determinar el número óptimo de clústeres (\(K\)) en el algoritmo K-Means, identificando visualmente un punto de inflexión ("codo")

```python
from sklearn.cluster import KMeans

inertia=[]
sil=[]

for x in range(2,11):
  valores=data
  Algoritmo= KMeans(n_clusters=x, max_iter=20,n_init=5)

  Grupos=Algoritmo.fit(valores)

  valores['cluster']=Grupos.labels_
  print(valores['cluster'].value_counts())
  sil.append(metrics.silhouette_score(valores[['x','y']], valores['cluster'],metric='euclidean'))
  inertia.append(Algoritmo.inertia_)

plt.plot(range(2, 11), inertia, marker='o', color='green')
plt.title('Método del codo') 
plt.xlabel('Número de clusters')
plt.ylabel('Inertia')
plt.show()

plt.bar(range(2, 11), sil)
plt.title('Silhouette')
plt.xlabel('Número de clusters')
plt.ylabel('Silhouette')
```

El número óptimo de grupos queda entre 3 y 4
