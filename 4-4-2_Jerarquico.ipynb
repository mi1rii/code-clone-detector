## Comparáción de tipos de enlace (linkage)
El _linkage_ se utiliza en el agrupamiento jerárquico para definir la distancia o similitud entre dos grupos diferentes. Esto decide qué grupos deben unirse a medida que se forma el dendrograma. Los distintos valores de linkage producen agrupamientos diferentes porque cambian la forma de medir la distancia entre clusters. Los métodos más comunes son:

-_single_:Usa la distancia mínima entre puntos. Suele generar el efecto chaining (clusters alargados).

-_complete_: Usa la distancia máxima. Produce grupos más compactos.

-_average_: Usa la distancia promedio entre elementos.

-_ward_: Minimiza la varianza interna de los clusters. Generalmente genera clusters balanceados.

```python
np.random.seed(42)

# Centros diseñados para provocar diferencias entre linkages
centros = [
    [0, 0],
    [4, 4],
    [8, 0]
]
Data, y = make_blobs(
    n_samples=180,
    centers=centros,
    cluster_std=[1.8, 1.0, 2.5],
    random_state=42
)

# Agregar algunos puntos "puente"
ruido = np.array([
    [2, 2],
    [3, 2.5],
    [5, 2.5],
    [6, 2]
])

Data = np.vstack([Data, ruido])

Data = pd.DataFrame(
    Data,
    columns=["x", "y"]
)

plt.scatter(Data['x'],Data['y'])
plt.show()
```

```python
link=['single','complete','average','ward']
Grupos={}
Sil={}

clusters=3

for x in link:
  Algoritmo = AgglomerativeClustering(n_clusters=clusters, linkage=x)
  Grupo=Algoritmo.fit(Data)
  print("------- ",x,"--------\n",Grupo.labels_)
  Sil[x]=metrics.silhouette_score(Data, Grupo.labels_,metric='euclidean')
  Grupos[x]=Grupo.labels_

  linkage_matrix = linkage(Data, method=x)
  dendrogram(linkage_matrix)
  plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for ax, metodo in zip(axes.ravel(), link):

    ax.scatter(
        Data["x"],
        Data["y"],
        c=Grupos[metodo],     # clusters desde el diccionario
        cmap="Set1",
        s=50
    )

    ax.set_title(metodo+' ('+str(round(Sil[metodo],3))+')')
    ax.set_xlabel("x")
    ax.set_ylabel("y")

# Ajustar espacios
plt.tight_layout()

# Mostrar figura
plt.show()
```
