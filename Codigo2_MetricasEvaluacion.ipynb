# Métricas de evaluación
Se trabajará con los siguientes datos ejemplo, dando por hecho que la lista "hipotesis" es el resultado de un modelo de clasificación, mientras que la lista "real" es la clase a la que pertenecen las instancias.

```python
#Clase real
real=['Pos','Pos','Neg','Neg','Neg','Pos','Pos','Pos','Pos','Pos','Neg','Pos','Neg','Pos']

#Clase asignada por el modelo
hipotesis=['Pos','Neg','Neg','Neg','Pos','Pos','Pos','Neg','Neg','Pos','Pos','Neg','Neg','Neg']

for x, y in zip(real, hipotesis):
    print(x,"-->",y)
```

## Matriz de confusión
Se puede obtener en formato texto, o bien, mediante una imagen

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

print("Matriz de confusión")
cm=confusion_matrix(real, hipotesis, labels=['Pos','Neg'])#(Primero la etiqueta de la clase que me interesa))
print(cm)
print(cm[0,0])#TP
print(cm[0,1])#FN
print(cm[1,0])#FP
print(cm[1,1])#TN


ConfusionMatrixDisplay.from_predictions(real,hipotesis,labels=['Pos','Neg'],cmap='gnuplot2_r')
plt.show()
```

## Precision, recall y F1
Se utiliza una función por métrica. Cuando nos interesa una sola clase, las funciones toman como parámetro la clase real, la hipótesis y la clase que nos interesa (objetivo). Cuando nos iteresan las dos clases, no se agrega la clase objetivo pero se utiliza el parámetro _average_ para determinar si es promedio simple (_macro_) o ponderado (_weighted_)

```python
#Metricas
from sklearn.metrics import accuracy_score,  precision_score, recall_score, f1_score


print("Exactitud: ",accuracy_score(real,hipotesis))

ClaseObjetivo='Pos'

print("Pos")
print("Precision: ",precision_score(real,hipotesis,pos_label=ClaseObjetivo))
print("Recall: ",recall_score(real,hipotesis,pos_label=ClaseObjetivo))
print("F1: ",f1_score(real,hipotesis,pos_label=ClaseObjetivo))

print("Neg")
ClaseObjetivo='Neg'
print("Precision: ",precision_score(real,hipotesis,pos_label=ClaseObjetivo))
print("Recall: ",recall_score(real,hipotesis,pos_label=ClaseObjetivo))
print("F1: ",f1_score(real,hipotesis,pos_label=ClaseObjetivo))


print("\n-------macro--------")
print("Precision: ",precision_score(real,hipotesis,average='macro'))
print("Recall: ",recall_score(real,hipotesis,average='macro'))
print("F1: ",f1_score(real,hipotesis,average='macro'))


print("\n---------weighted-----------")
print("Precision: ",precision_score(real,hipotesis,average='weighted'))
print("Recall: ",recall_score(real,hipotesis,average='weighted'))
print("F1: ",f1_score(real,hipotesis,average='weighted'))
```

Una manera más sencilla de obtener todas las métricas es el reporte de clasificación, que da toda la información ya sea en formato cadena o en un diccionario. El diccionario se genera con el parámetro _output_dict=True_

```python
#Reporte de clasificación
from sklearn.metrics import classification_report

""" El reporte de clasificación """
reporte=classification_report(real,hipotesis,labels=['Pos','Neg'])
print(reporte)

print(type(reporte))

reporte=classification_report(real,hipotesis,labels=['Pos','Neg'], output_dict=True)
print(reporte['accuracy'])
print(reporte['Pos']['recall'])
print(reporte['macro avg']['precision'])
print(reporte['macro avg']['f1-score'])
```
