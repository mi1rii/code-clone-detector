# Arbol de decisión
## Lectura de datos

```python
import pandas as pd
from pandas import Series,DataFrame

setInicial=pd.read_csv('../Datos/IrisBinario.csv')

#Recuperar nombres de las columnas de los atributos y la clase
atributosName=setInicial.columns[:-1].to_list()
atributoClase=setInicial.columns[-1]
print("Nombre de atributos:",atributosName)
print("Atributos objetivo:",atributoClase)

#Datos de los atributos
dataGral=setInicial[atributosName]

#Datos de la clase (vector de elementos reales)
claseGral=setInicial[atributoClase]
nombreClases=claseGral.value_counts().index.to_list()

#Instancias por clase
print("Clases en el conjunto de datos:", nombreClases)
print("\nInstancias por clase: ", claseGral.value_counts())
```

```python
#Visualización de datos
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

#primero se pasa la clase a enteros, para asignar un mapa
clases = claseGral.astype("category")
color = clases.cat.codes

plt.figure(figsize=(14, 14))
scatter_matrix(
    dataGral,
    figsize=(14, 14),
    diagonal='hist',
    alpha=0.5,
    c=color
)
plt.show()
```

## Implementación del algoritmo
### Entrenamiento
Se implementa el algoritmo con la función _DecisionTreeClassifier()_. Algunos de los parámetros más utilizados son:
* _criterion_: gini, entropy (gini)
* _max_depth_: número máximo de profundidad (infinito)
* _min_samples_split_: mínimo de instancias para dividir un nodo (2)
* _min_samples_leaf_: mínimo de instancias para ser una hoja (1)

Para entrenar el modelo, se utiliza la función _.fit()_ la cual toma los datos y la clase para extraer los patrones. En este algoritmo, se genera un árbol que puede imprimirse a modo de imagen o texto. Además, se pueden obtener otros datos de importancia

```python
#Librería para la implementación dle algoritmo
from sklearn.tree import DecisionTreeClassifier

################## Determinar el algoritmo ######################
modelo = DecisionTreeClassifier(max_depth=3, criterion='gini')

################## Entrenar el modelo ###########################
modelo.fit(dataGral, claseGral)

############### Imprimir el modelo #########
#Esto solo es para árbol de decisión, cada algoritmo tiene sus propios métodos para visualizar el modelo generado

#Profundidad maxima
profundidadMax = modelo.get_depth()
print(f"Profundidad total del árbol:",profundidadMax,'\n\n')

#Arbol en texto
from sklearn.tree import export_text
arbol=export_text(modelo, feature_names=atributosName)#, max_depth=3)
print(arbol)



#Arbol en imagen
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(modelo,feature_names=atributosName,class_names=nombreClases)#, max_depth=2,filled=True)
plt.show()
```
