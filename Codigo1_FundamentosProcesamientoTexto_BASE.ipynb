## Lemas y POS Tagging
Categorías más utilizadas.
* ADJ: Adjetivo (ej. bonito, azul).
* ADP: Adposición (preposiciones, ej. a, con, de).
* ADV: Adverbio (ej. muy, rápidamente).
* AUX: Verbo auxiliar (ej. ser, estar, haber).
* CONJ/CCONJ/SCONJ: Conjunciones (coordinantes y subordinantes, ej. y, pero, que).
* DET: Determinante (ej. el, una, este).
* INTJ: Interjección.
* NOUN: Sustantivo/Nombre (ej. perro, mesa).
* NUM: Numeral (ej. tres, 15).
* PART: Partícula.PRON: Pronombre (ej. yo, se, quien).
* PROPN: Nombre propio (ej. María, Madrid).
* PUNCT: Puntuación (ej. ., ¿, ,).
* SCONJ: Conjunción subordinante.
* SYM: Símbolo.
* VERB: Verbo (ej. correr, escribir)

## Tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer

VectorTfIdf = TfidfVectorizer()
tfidf = VectorTfIdf.fit_transform(Lemas)
atributos=VectorTfIdf.get_feature_names_out()
dfTfidf = pd.DataFrame(tfidf.toarray(), columns=atributos)
dfTfidf.round(3)

```

## Similitud entre textos
Similitud coseno: Se puede utilizar el conteo o el tfidf

```python
from sklearn.metrics.pairwise import cosine_similarity

similitudCoseno = cosine_similarity(dfTfidf)

dfCoseno = pd.DataFrame(similitudCoseno, index=Lemas, columns=Lemas)
dfCoseno
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(dfCoseno, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Similitud entre textos (Coseno TF-IDF)", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
